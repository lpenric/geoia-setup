```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
# Données de test : Paris et Lyon
df = pd.DataFrame({
    "ville": ["Paris", "Lyon"],
    "lat": [48.85, 45.75],
    "lon": [2.35, 4.85]
})
print(df)
```

```python
# Conversion en GeoDataFrame avec coordonnées géographiques
gdf = gpd.GeoDataFrame(
    df,
    geometry=gpd.points_from_xy(df.lon, df.lat),
    crs="EPSG:4326"  # Système de coordonnées WGS84 (standard GPS)
)
print(gdf.head())
```

```python
# Tracé de la carte
fig, ax = plt.subplots(figsize=(8, 6))
gdf.plot(ax=ax, marker='o', color='red', markersize=50)
plt.title("Carte test : Paris et Lyon")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.grid(True)

# Sauvegarde de la carte
plt.savefig("map_test.png", bbox_inches='tight', dpi=150)
plt.show()
```
